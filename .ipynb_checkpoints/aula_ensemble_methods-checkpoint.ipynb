## Árvores de Decisão

Árvores de decisão são algoritmos de aprendizagem supervisionada hierárquica.

Elas nos ajudam principalmente com:

- Classificação e regressão
- Modelagem não linear
- Quebrar dados por meio de decisões binárias

image.png

## 1.1 [DecisionTreeClassifier](https://scikit-learn.org/stable/modules/generated/sklearn.tree.DecisionTreeClassifier.html)

```python
from sklearn.datasets import load_iris
import pandas as pd
import numpy as np

iris = load_iris()
data = pd.DataFrame(data= np.c_[iris['data'], iris['target']],
                     columns= iris['feature_names'] + ['target'])
data.drop(columns=['sepal length (cm)', 'sepal width (cm)'], inplace=True)
data
```

```python
X = data.drop(columns=['target']).values
y = data.target.values
```

```python
# Instantiate and train model
from sklearn.tree import DecisionTreeClassifier
tree_clf = DecisionTreeClassifier(max_depth=2, random_state=2)
tree_clf.fit(X,y)
```

```python
import graphviz

# Export model graph
from sklearn.tree import export_graphviz
export_graphviz(tree_clf, out_file="iris_tree.dot",
                feature_names=data.drop(columns=['target']).columns,
                class_names=['0','1','2'],
                rounded=True, filled=True)

# Import model graph
with open("iris_tree.dot") as f:
    dot_graph = f.read()
    display(graphviz.Source(dot_graph))
```

## Index Gini

O índice de Gini mede a capacidade de cada característica de separar os dados.

Ele calcula a impureza de cada nó, entre [0, 1]; quanto menor, melhor.

image.png

*pi* sendo a razão entre as observações em uma classe, *i* o número total de observações restantes em um determinado nó:

image.png

```python
# Calculate gini of root node
1 - (50/150)**2 - (50/150)**2 - (50/150)**2
```

```python
# Calcultate gini green leaf
1 - 0**2 - (49/54)**2 - (5/54)**2
```

## Como uma árvore cresce?

A estrutura da árvore é definida através dos seguintes passos:

1. Comece no nó raiz, que contém todo o seu conjunto de dados
2. Experimente várias combinações de combos (recurso, limiar); cada um dividiria seu conjunto de dados em 2 nós filhos
3. Para cada combinação, calcule o índice Gini médio ponderado de ambos os nós filhos (ponderado pelo número de instâncias)
4. Selecione a combinação (recurso, limite) que produz o índice mais baixo (ou seja, os "nós filhos mais puros")
5. Divida o conjunto de dados em dois usando esta regra
6. Repita a etapa 2 para ambos os subconjuntos
7. Pare quando nenhum recurso melhorar a impureza do nó (com que risco?)

[Source: CART algorithm (Aurelien Geron)](https://wagon-public-datasets.s3.amazonaws.com/data-science-images/lectures/machine-learning/tree_growth_aurelien_geron.png)

### Predição

- Um novo ponto é passado pela árvore de cima para baixo até chegar a uma folha
- A classe mais representada nessa folha é a classe prevista para um determinado ponto de dados

*Exemplo*: considere o seguinte ponto de dados: `X_new = [4(comprimento), 1(largura)]`

image.png

```python
# Let's predict the class of a new flower
print(tree_clf.predict([[4,1]]))
```

```python
# Predict proba
print(tree_clf.predict_proba([[4,1]]))
```

❗️90% não é uma "probabilidade" per se, corresponde à proporção de classes no nó.

As árvores não são classificadores calibrados (em oposição aos regressores logísticos)

## Pense nas árvores de decisão como classificadores "ortogonais"

```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.colors import ListedColormap

def plot_decision_regions(X, y, classifier, resolution=0.02):
    # Configurando marcadores e cores
    markers = ('s', 'x', 'o', '^', 'v')
    colors = ('red', 'blue', 'lightgreen', 'gray', 'cyan')
    cmap = ListedColormap(colors[:len(np.unique(y))])

    # Determinando limites do gráfico
    x1_min, x1_max = X[:, 0].min() - 1, X[:, 0].max() + 1
    x2_min, x2_max = X[:, 1].min() - 1, X[:, 1].max() + 1

    # Criando uma grade de pontos com resolução
    xx1, xx2 = np.meshgrid(np.arange(x1_min, x1_max, resolution),
                           np.arange(x2_min, x2_max, resolution))

    # Classificando cada ponto na grade
    Z = classifier.predict(np.array([xx1.ravel(), xx2.ravel()]).T)
    Z = Z.reshape(xx1.shape)

    # Plotando a superfície de decisão
    plt.contourf(xx1, xx2, Z, alpha=0.3, cmap=cmap)
    plt.xlim(xx1.min(), xx1.max())
    plt.ylim(xx2.min(), xx2.max())

    # Plotando exemplos de treinamento
    for idx, cl in enumerate(np.unique(y)):
        plt.scatter(x=X[y == cl, 0],
                    y=X[y == cl, 1],
                    alpha=0.8,
                    c=[cmap(idx)],
                    marker=markers[idx],
                    label=cl)

    plt.legend(loc='upper left')
    plt.show()

# Exemplo de uso:
# plot_decision_regions(X, y, classifier=tree_clf)
```

```python
# Hand-made function to plot all predictions
plot_decision_regions(X, y, classifier=tree_clf)
```
